```python
x = np.linspace(0, 1, 1000)

# Calculate the probability density function (PDF) for the Beta distribution
pdf = scipy.stats.beta.pdf(x, 2, 8)

# Create the plot
plt.figure()
plt.plot(x, pdf, label=f'Beta(2, 8)', color='blue')
plt.title(f'Beta Distribution (alpha=2, beta=8)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.integrate import quad

x = np.linspace(0, 1, 100)
pdf_values = []
def pdf_sum(x_val):
    return stats.beta.pdf(x_val, 2, 8) + stats.beta.pdf(x_val, 8, 2)

# Compute the PDF values at each point using quad
for x_i in x:
    pdf_values.append(quad(pdf_sum, 0, x_i))
# pdf_values = [quad(pdf_sum, 0, xi)[0] for xi in x]

# Generate 10,000 samples from each Beta distribution
samples1 = np.random.beta(2, 8, 10000)
samples2 = np.random.beta(8, 2, 10000)

# Sum the samples from the two distributions
sum_samples = samples1 + samples2

# Create a histogram of the summed samples
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf_values, label="PDF")
plt.title("PDF of Beta(2, 8) + Beta(8, 2)")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(sum_samples, bins=50, density=True, alpha=0.5, color='b', label="Sampled PDF")
plt.title("Histogram of Sampled Data (Beta(2, 8) + Beta(8, 2))")
plt.xlabel("Sum of Samples")
plt.ylabel("Normalized Frequency")
plt.legend()

plt.tight_layout()
plt.show()
```
